## Batch Data

```python
def calc(weights,batches,biases):
  return np.dot(batches,np.transpose(weights))+biases 
```

```python
inputs = [[1.0, 2.0, 3.0, 2.5],
 [2.0, 5.0, -1.0, 2.0],
 [-1.5, 2.7, 3.3, -0.8]]
weights = [[0.2, 0.8, -0.5, 1.0],
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]
biases = [2.0, 3.0, 0.5]
output=calc(weights,inputs,biases)
print(output)
```
