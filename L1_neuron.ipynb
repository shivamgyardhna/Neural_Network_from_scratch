## Multi Layer


```python
def res(weight,input,bias):
  res=0
  for i in range(len(weight)):
    res+=weight[i]*input[i]
  res+=bias
  return res 
```

```python
inputs = [1, 2, 3, 2.5]
weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]
wt=[weights1,weights2,weights3]
bias1 = 2
bias2 = 3
bias3 = 0.5
bias=[bias1,bias2,bias3]
output=[]
neuron=3
for i in range(neuron):
  output.append(res(wt[i],inputs,bias[i]))
print(output)  
```
